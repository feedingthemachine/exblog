```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
import warnings
warnings.simplefilter('ignore')
plt.rcParams['figure.figsize'] = (15,80)
```

```python
df = pd.read_csv('chennai_reservoir_levels.csv').drop('Date',axis=1)
```

```python
df.head()
```

```python
for index , (colname,serie) in enumerate(df.iteritems()):
    plt.subplot(31,1, index+1)
    if pd.api.types.is_float_dtype(serie) is True:
        sns.distplot(serie,fit=norm)
        plt.axvline(np.mean(serie), color='slategrey')
    elif pd.api.types.is_integer_dtype(serie) is True:
        sns.countplot(serie)
    plt.title(colname, fontsize=16)
    plt.xlabel(''); plt.ylabel('');plt.tight_layout()
```

```python
df_scaled = StandardScaler()
df_scaled = pd.DataFrame(df_scaled.fit_transform(df), columns = df.columns) 
df_scaled.head()
```

```python
for index , (colname,serie) in enumerate(df_scaled.iteritems()):
    plt.subplot(31,1, index+1)
    if pd.api.types.is_float_dtype(serie) is True:
        sns.distplot(serie,fit=norm)
        plt.axvline(np.mean(serie), color='slategrey')
    elif pd.api.types.is_integer_dtype(serie) is True:
        sns.countplot(serie)
    plt.title(colname, fontsize=16)
    plt.xlabel(''); plt.ylabel('');plt.tight_layout()
```
